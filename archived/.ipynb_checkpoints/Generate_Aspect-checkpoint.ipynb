```python
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_excel('../data/original/QBO_IOS_With_Aspect.xlsx', sheet_name='Review Text')
```

```python
df.head()
```

```python
def add_comma(aspect):
    comma = ','
    if aspect != '':
        aspect += comma

    return aspect

def add_aspect():
    for index, row in df.iterrows():
        if pd.isnull(row['Aspect']):
            aspect = ''

            if "update" in row['Subject'] or "update" in row['Body']:
                aspect += "updates"

            if "ios" in row['Subject'] or "ios" in row['Body'] or "ipad" in row['Subject'] or "ipad" in row['Body']:
                aspect = add_comma(aspect)
                aspect += "compatibility"

            if "crash" in row['Subject'] or "crash" in row['Body']:
                aspect = add_comma(aspect)
                aspect += "durability"

            if "crash" in row['Subject'] or "crash" in row['Body']:
                aspect = add_comma(aspect)
                aspect += "durability"
            
            if "transact" in row['Subject'] or "transact" in row['Body']:
                aspect = add_comma(aspect)
                aspect += "transaction"
            
            if aspect != '':
                print(aspect)
```

```python
add_aspect()
```

```python
millis = str(int(round(time.time() * 1000)))
df.to_excel("../data/QBO_IOS_With_Auto_Aspect_" + millis + ".xlsx", sheet_name='Review Text') 
```
