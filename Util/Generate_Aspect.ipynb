```python
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_excel('../data/reviewtosentences/QBO_IOS_ReviewToSentences.xlsx', sheet_name='Review Text')
```

```python
df.head()
```

```python
def add_comma(aspect):
    comma = ','
    if aspect != '':
        aspect += comma

    return aspect

def found_match_and_replace(dfrow, keyword, aspect, curr_aspect_list):
    if keyword in dfrow['Subject'].lower() or keyword in dfrow['Body'].lower():
        curr_aspect_list = add_comma(curr_aspect_list)
        curr_aspect_list += aspect
        
    return curr_aspect_list

def add_aspect(df_original):
    count = 0
    
    for index, row in df_original.iterrows():

        aspect = ''

        aspect = found_match_and_replace(row, "crash", "durability", aspect)
        aspect = found_match_and_replace(row, "payroll", "payroll", aspect)
        aspect = found_match_and_replace(row, "pay", "payment", aspect)
        aspect = found_match_and_replace(row, "bank", "banking", aspect)

        aspect = found_match_and_replace(row, "report", "report", aspect)

        aspect = found_match_and_replace(row, "mileage", "mileage", aspect)
        aspect = found_match_and_replace(row, "account", "accounting", aspect)
        aspect = found_match_and_replace(row, "expense", "expenses", aspect)
        aspect = found_match_and_replace(row, "subscription", "subscription", aspect)
        aspect = found_match_and_replace(row, "attachment", "attachment", aspect)
        aspect = found_match_and_replace(row, "dashboard", "dashboard", aspect)
        aspect = found_match_and_replace(row, "currency", "currency", aspect)
        aspect = found_match_and_replace(row, "vendor", "vendor", aspect)
        aspect = found_match_and_replace(row, "paperwork", "paperwork", aspect)
        aspect = found_match_and_replace(row, "print", "print", aspect)

        for item in ['receipt', 'capture']:
            aspect = found_match_and_replace(row, item, "receipt capture", aspect)
            if "receipt capture" in aspect:
                break

        for item in ['ios', 'iphone', 'ipad', 'apple', 'launch']:
            aspect = found_match_and_replace(row, item, "compatibility", aspect)
            if "compatibility" in aspect:
                break

        for item in ['update', 'version']:
            aspect = found_match_and_replace(row, item, "updates", aspect)
            if "updates" in aspect:
                break

        for item in ['sale', 'tax']:
            aspect = found_match_and_replace(row, item, "sales tax", aspect)
            if "sales tax" in aspect:
                break

        for item in ['transact', 'transf', 'deposite', 'credit card']:
            aspect = found_match_and_replace(row, item, "transaction", aspect)
            if "transaction" in aspect:
                break 

        for item in ['FAQ', 'help', 'tutorial', 'video']:
            aspect = found_match_and_replace(row, item, "help", aspect)
            if "help" in aspect:
                break 

        for item in ['bill', 'invoic']:
            aspect = found_match_and_replace(row, item, "invoicing", aspect)
            if "invoicing" in aspect:
                break 

        for item in ['call', 'hour', 'minute', 'agent', 'representative', 'email', 'screenshot', 'logged', 'desk']:
            aspect = found_match_and_replace(row, item, "customers service", aspect)
            if "customers service" in aspect:
                break            

        for item in ['rate', 'price', 'pricing', 'prices', 'cheap', 'expensive', 'fee', 'refund', 'money', 'dollar', '$']:
            aspect = found_match_and_replace(row, item, "price", aspect)
            if "price" in aspect:
                break

        for item in ['sync', 'integration', 'online', 'qbo', 'export', 'import']:
            aspect = found_match_and_replace(row, item, "integration", aspect)
            if "integration" in aspect:
                break

        for item in ['design', 'catego', 'find', 'menu', 'naviga', 'button', 'responsive']:
            aspect = found_match_and_replace(row, item, "navigation", aspect)
            if "navigation" in aspect:
                break 

        for item in ['record', 'result', 'data']:
            aspect = found_match_and_replace(row, item, "data", aspect)
            if "data" in aspect:
                break

        for item in ['slow', 'fast', 'load', 'perform']:
            aspect = found_match_and_replace(row, item, "performance", aspect)
            if "performance" in aspect:
                break

        for item in ['global', 'international', 'india', 'mexico', 'australia', 'philippines', 'canada', ' UK ']:
            aspect = found_match_and_replace(row, item, "global", aspect)
            if "global" in aspect:
                break

        for item in ['client', 'customer']:
            aspect = found_match_and_replace(row, item, "customers", aspect)
            if "customers" in aspect:
                break

        if aspect == '':
            aspect = "general,quality"

        df_original.at[index, 'Aspect'] = aspect
            
    return df_original
```

```python
df_generated_aspect = add_aspect(df)
```
